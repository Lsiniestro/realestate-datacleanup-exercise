```python
from ipyleaflet import Map, basemaps

# Mapa centrado en (60 grados latitud et -2.2 grados longitud)
# Latitud, longitud


nuevo_ds['latitude'] = nuevo_ds['latitude'].astype("string")
nuevo_ds['longitude'] = nuevo_ds['longitude'].astype("string")
nuevo_ds['latitude'] = nuevo_ds['latitude'].replace(',', '.', regex=True)
nuevo_ds['longitude'] = nuevo_ds['longitude'].replace(',', '.', regex=True)

map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
## Aquí: traza la coordenadas de los estados

from ipyleaflet import Marker

## PON TU CÓDIGO AQUÍ:
coordinates = nuevo_ds[['latitude', 'longitude']].values
populations = nuevo_ds['level5'].values
test = True
# Adding Markers to the map with Popup
for (lat, lon), pop in zip(coordinates, populations):
    if lat != '0' and lon != '0':
        marker = Marker(location=(lat, lon))
        map.add_layer(marker)
    else:
        continue

map
```
